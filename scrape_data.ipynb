```python
genre_link= ["https://www.goodreads.com/genres/fiction","https://www.goodreads.com/genres/romance", "https://www.goodreads.com/genres/mystery", "https://www.goodreads.com/genres/horror", "https://www.goodreads.com/genres/fantasy", "https://www.goodreads.com/genres/science-fiction", "https://www.goodreads.com/genres/historical-fiction", "https://www.goodreads.com/genres/thriller", "https://www.goodreads.com/genres/poetry", "https://www.goodreads.com/genres/crime", "https://www.goodreads.com/genres/young-adult", "https://www.goodreads.com/genres/science", "https://www.goodreads.com/genres/self-help", "https://www.goodreads.com/genres/non-fiction"]
```

```python
books=[]
for links in genre_link:
  response = requests.get(links, headers=headers)
  soup = BeautifulSoup(response.text, 'html.parser')


  for a_tag in soup.find_all('a', href=True):
    href= a_tag['href']
    if href.startswith('/book/show/'):
      full_link= "https://www.goodreads.com" + href.split('?')[0]
      if full_link not in books:
        books.append(full_link)


print(books)
```

```python
len(books)
```

```python
from __future__ import print_function
title=[]
author=[]
rating=[]
genre=[]
review1=[]
review2=[]
review3=[]


for url in books:
  response= requests.get(url, headers=headers)
  soup=BeautifulSoup(response.text, 'html.parser')


  title.append(soup.find('h1', class_='Text Text__title1').text.strip())
  author.append(soup.find('h3', class_='Text Text__title3 Text__regular').text.strip())
  rating.append(soup.find('div', class_='RatingStatistics__rating').text.strip())

  review_block= soup.find_all('section', class_='ReviewText')
  top_reviews=[block.find('span', class_ ='Formatted').text.strip() for block in review_block[:3]]

  review1.append(top_reviews[0] if len(top_reviews)>0 else 'NA')
  review2.append(top_reviews[1] if len(top_reviews)>1 else 'NA')
  review3.append(top_reviews[2] if len(top_reviews)>2 else 'NA')

data= {'title': title, 'author':author, 'rating':rating, 'review1':review1, 'review2': review2, 'review3':review3}

```

```python
df=pd.DataFrame(data)
df
```
